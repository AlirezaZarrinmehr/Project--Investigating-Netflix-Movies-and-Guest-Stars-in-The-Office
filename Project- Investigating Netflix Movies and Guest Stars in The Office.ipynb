<p>In this notebook, we will take a look at a dataset of The Office episodes, and try to understand how the popularity and quality of the series varied over time. To do so, we will use the following dataset: <code>datasets/office_episodes.csv</code>, which was downloaded from Kaggle <a href="https://www.kaggle.com/nehaprabhavalkar/the-office-dataset">here</a>.</p>
<p>This dataset contains information on a variety of characteristics of each episode. In detail, these are:

```python
import pandas as pd
import matplotlib.pyplot as plt

#read the dataset
office_episodes = pd.read_csv("datasets/office_episodes.csv")

#define colors based on the scaled_ratings column
colors = office_episodes["scaled_ratings"].apply(lambda x: "red" if x <0.25 else "orange" if x>=0.25 and x<0.5 else "lightgreen" if x>=0.5 and x <0.75 else "darkgreen")

#define size based on the presence of guest stars
size = office_episodes["has_guests"].apply(lambda x : 250 if x==True else 25)

#create a figure and axis object
fig, ax = plt.subplots()

#plot a scatter plot of episode number vs viewership, using the colors and size we defined
ax.scatter(office_episodes["episode_number"], office_episodes["viewership_mil"], c = colors, s = size)

#set the title and labels for the plot
ax.set_title("Popularity, Quality, and Guest Appearances on the Office")
ax.set_xlabel("Episode Number")
ax.set_ylabel("Viewership (Millions)")

#set the figure size
plt.rcParams['figure.figsize'] = [11, 7]

#clear the old plots

plt.clf()

#show the plot
plt.show()

#office_episodes["guest_stars"][office_episodes["viewership_mil"]==office_episodes["viewership_mil"].max()]
top_star = "Jessica Alba"
```
